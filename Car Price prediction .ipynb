### Conclusion

 We can remove the columns which of no use for our analysis

```python
final_data = dataset[['price','enginetype','fueltype', 'aspiration','carbody','cylindernumber', 'drivewheel','wheelbase','curbweight', 'enginesize', 'boreratio','horsepower', 'citympg','highwaympg', 'carlength','carwidth']]
```

<h1><center> Data Preprocessing </center></h1>

```python
final_data.info()
```

```python
final_data.dtypes.value_counts()
```

we have 6 catagorical features to deal with for that we will use Encoding

### Encoding

```python
from sklearn.preprocessing import LabelEncoder
categorical=['enginetype','fueltype','aspiration','carbody','cylindernumber','drivewheel']
for x in categorical:
    encoder = LabelEncoder()
    final_data[x]= encoder.fit_transform(final_data[x])
final_data.head()
```

### Splitting the Data

```python
train_data = final_data.drop('price',axis=1)
target_data = final_data['price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_data, target_data, test_size=0.2, random_state=50)
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

###  Building Models

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model = LinearRegression()
model.fit(X_train,y_train)
preds=model.predict(X_test)
print(r2_score(preds,y_test))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train,y_train)
preds=model.predict(X_test)
print(r2_score(preds,y_test))
```

### Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train,y_train)
preds=model.predict(X_test)
print(r2_score(preds,y_test))
```

### KNeighbours Regressor

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=2)
model.fit(X_train,y_train)
preds=model.predict(X_test)
print(r2_score(preds,y_test))
```

# Conclusion

we finally got an r2 score of 91% using Random Forest
